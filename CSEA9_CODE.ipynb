```python
months = list()
months.append(df[(df['Date'] >= '2020-01-01') & (df['Date'] <= '2020-01-15')])
months.append(df[(df['Date'] >= '2020-01-16') & (df['Date'] <= '2020-01-31')])
months.append(df[(df['Date'] >= '2020-02-01') & (df['Date'] <= '2020-02-15')])
months.append(df[(df['Date'] >= '2020-02-16') & (df['Date'] <= '2020-02-29')])
months.append(df[(df['Date'] >= '2020-03-01') & (df['Date'] <= '2020-03-15')])
months.append(df[(df['Date'] >= '2020-03-16') & (df['Date'] <= '2020-03-31')])
months.append(df[(df['Date'] >= '2020-04-01') & (df['Date'] <= '2020-04-15')])
months.append(df[(df['Date'] >= '2020-04-15') & (df['Date'] <= '2020-04-30')])
```

```python
statewise = pd.DataFrame(columns = df.columns)
for month, name in zip(months, month_names):
    for state in states:
        temp = month[(month['Name of State / UT'] == state)]
        statewise = statewise.append(temp[temp['Total Confirmed cases'] == temp['Total Confirmed cases'].max()])
    statewise = statewise.drop_duplicates(subset = "Name of State / UT", keep = 'last')
    statewise = statewise[['Name of State / UT', 'Total Confirmed cases']]
    statewise.to_csv(str(str(name) + '_output.csv'), index=False)
```

```python
import pandas as pd
import numpy as np
states = np.genfromtxt('states.csv', delimiter=',', dtype=str)
month_names = ['jan_1', 'jan_2', 'feb_1', 'feb_2', 'mar_1', 'mar_2', 'apr_1', 'apr_2']
month_inputs = list()
for month in month_names:
    temp = pd.read_csv(month + '_output.csv')
    month_inputs.insert(month_names.index(month), temp)
for month in month_names:
    adj = np.zeros((len(states), len(states)), dtype=int)
    arr = np.array(month_inputs[month_names.index(month)]['Name of State / UT'])
    for elem in arr:
        for i in range(len(states)):
            if elem == states[i]:
                j = int(np.where(states == elem)[0])
                for k in range(j+1):
                    adj[k][j] = month_inputs[month_names.index(month)].iloc[j]['Total Confirmed cases']
                    adj[j][k] = month_inputs[month_names.index(month)].iloc[j]['Total Confirmed cases']
                break
    np.savetxt(month + "_adjacency.csv", adj, delimiter=',', fmt='%s')
```

```python
import csv
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

with open('jan_2_matrix.csv', 'r') as file:
    reader = csv.reader(file)
    header = next(reader)
    for row in reader:
        source, destination = row[0], row[1]
        G.add_edge(source, destination)

pagerank = nx.pagerank(G)
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)


print("Node\tPageRank Centrality")
for node, centrality in pagerank.items():
    print(f"{node}\t{centrality}")
    
print("\nDegree centrality:")
for node, centrality in degree_centrality.items():
    print(f"{node}\t{centrality}")

print("\nBetweenness centrality:")
for node, centrality in betweenness_centrality.items():
    print(f"{node}\t{centrality}")

print("\nCloseness centrality:")
for node, centrality in closeness_centrality.items():
    print(f"{node}\t{centrality}")
    

# Find the node with the highest PageRank value
highest_pr_node = max(pagerank, key=pagerank.get)

# Find the node with the maximum degree centrality
highest_deg_centrality_node = max(degree_centrality, key=degree_centrality.get)

# Find the node with the maximum betweenness centrality
highest_bet_centrality_node = max(betweenness_centrality, key=betweenness_centrality.get)

# Find the node with the maximum closeness centrality
highest_clo_centrality_node = max(closeness_centrality, key=closeness_centrality.get)

# Display the results
print(f"\n\nThe highest PageRank value is {pagerank[highest_pr_node]} for node {highest_pr_node}.")
print(f"\n\nThe maximum degree centrality is {degree_centrality[highest_deg_centrality_node]} for node {highest_deg_centrality_node}.")
print(f"\n\nThe maximum betweenness centrality is {betweenness_centrality[highest_bet_centrality_node]} for node {highest_bet_centrality_node}.")
print(f"\n\nThe maximum closeness centrality is {closeness_centrality[highest_clo_centrality_node]} for node {highest_clo_centrality_node}.")

nx.draw(G, with_labels=True, node_color=list(pagerank.values()), cmap=plt.get_cmap('viridis'), node_size=500, alpha=0.8, linewidths=0.5, edgecolors='gray')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('jan_2_matrix.csv')

# Group the data by the from and to locations and calculate the number of trips between each pair
heatmap_data = data.groupby(['0', '1']).size().reset_index(name='count')

# Create a pivot table with the from locations as rows and the to locations as columns
heatmap_data = heatmap_data.pivot(index='0', columns='1', values='count')

# Create the heatmap using Seaborn
sns.heatmap(heatmap_data, cmap='YlOrRd')

# Customize the plot
plt.title('Travel Frequency Heatmap')
plt.xlabel('To Location')
plt.ylabel('From Location')

plt.show()
```
